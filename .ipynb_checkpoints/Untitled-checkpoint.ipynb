```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv('test_Y3wMUE5_7gLdaTN.csv')
```

```python
df = df.dropna()
```

```python
df['Credit_History'] = np.where(df['Credit_History']==1,0,1)
```

```python
df_filtered = df[['Gender', 'Married', 'Dependents', 'Education', 'ApplicantIncome', 'LoanAmount',
       'Loan_Amount_Term', 'Property_Area','Credit_History']]
```

```python
df_filtered = df_filtered[df_filtered['Dependents'] != '3+']
```

```python
from sklearn.compose import ColumnTransformer
# Define columns to be one-hot encoded
columns_to_encode = ['Gender', 'Married', 'Education', 'Property_Area']

# Initialize a ColumnTransformer with the OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('one_hot', OneHotEncoder(sparse_output=False, drop='first'), columns_to_encode)
    ],
    remainder='passthrough'  # Leave the rest of the columns unchanged
)

# Fit and transform the data
df_encoded_array = preprocessor.fit_transform(df_filtered)

# Get feature names after one-hot encoding for creating DataFrame
feature_names = (preprocessor.named_transformers_['one_hot'].get_feature_names_out(columns_to_encode)
                 .tolist() + [col for col in df_filtered.columns if col not in columns_to_encode])

# Convert to DataFrame
df_encoded = pd.DataFrame(df_encoded_array, columns=feature_names)

# Convert all columns to integer
df_encoded = df_encoded.astype(int)
```
